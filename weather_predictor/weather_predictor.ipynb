```python
core_weather["target"] = core_weather.shift(-1)["temp_max"] 
# posiciona todas as linhas uma posição atrás, e cria coluna target com temp max do próximo dia
```

```python
from sklearn.linear_model import Ridge
# ridge regression minimiza overfitting

reg = Ridge(alpha=.1)
# alpha controla o quanto os coeficientes do modelo são penalizados, quanto maior
# a penalidade, maior a prevenção do overfitting
```

```python
predictors = ["precip", "temp_max", "temp_min"]
```

```python
train = core_weather.loc[:"1984-12-31"]
```

```python
test = core_weather.loc["1985-01-01":]
```

```python
reg.fit(train[predictors], train["target"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(test["target"], predictions)
```

```python
combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
combined.columns = ["actual", "predictions"]
```

```python
combined
```

```python
combined.plot()
```

```python
reg.coef_
# chuva possui impacto negativo na temperatura
```

```python
def create_predictions(predictors, core_weather, reg):
    train = core_weather.loc[:"1984-12-31"]
    test = core_weather.loc["1985-01-01":]
    reg.fit(train[predictors], train["target"])
    predictions = reg.predict(test[predictors])
    error = mean_absolute_error(test["target"], predictions)
    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ["actual", "predictions"]

    return error, combined
```

```python
core_weather["month_max"] = core_weather["temp_max"].rolling(30).mean()
# volta 30 dias a partir de cada dia e calcula a média,
```

```python
core_weather
```

```python
core_weather["month_day_max"] = core_weather["month_max"] / core_weather["temp_max"]
# o quão diferente é a temperatura máxima em um dia em relação à temperatura máxima naquele mês,
# informação pode ser útil ao algoritmo
```

```python
core_weather["max_min"] = core_weather["temp_max"] / core_weather["temp_min"] 
# relação entre temperatura máxima e mínima,
# também pode ser útil ao algoritmo
```

```python
predictors = ["precip", "temp_max", "temp_min", "month_max", "month_day_max", "max_min"]
```

```python
core_weather = core_weather.iloc[30:,:].copy()
```

```python
error, combined = create_predictions(predictors, core_weather, reg)
```

```python
error
```

```python
combined.plot()
```

```python
core_weather["monthly_avg"] = core_weather["temp_max"].groupby(core_weather.index.month).apply(lambda x: x.expanding(1).mean()).reset_index(level=0, drop=True)
# seleciona apenas linhas anteriores e calcula as médias
```

```python
core_weather
```

```python
core_weather["day_of_the_year_avg"] = core_weather["temp_max"].groupby(core_weather.index.day_of_year).apply(lambda x: x.expanding(1).mean()).reset_index(level=0, drop=True)
```

```python
predictors = ["precip", "temp_max", "temp_min", "month_max", "month_day_max", "max_min", "monthly_avg", "day_of_the_year_avg"]
```

```python
error, combined = create_predictions(predictors, core_weather, reg)
```
